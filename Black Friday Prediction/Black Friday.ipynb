```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
print(train.shape)
print(test.shape)
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['Product_Category_2'] = train['Product_Category_2'].fillna(train['Product_Category_2'].mean())
train['Product_Category_3'] = train['Product_Category_2'].fillna(train['Product_Category_3'].mean())
```

```python
test['Product_Category_2'] = test['Product_Category_2'].fillna(test['Product_Category_2'].mean())
test['Product_Category_3'] = test['Product_Category_2'].fillna(test['Product_Category_3'].mean())
```

```python
train.head()
```

```python
test.head()
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
df = pd.concat([train, test], ignore_index= True, sort= False)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Product_ID'] = df['Product_ID'].str.replace('P00','')
```

```python
df.Age.value_counts()
```

```python
def age(x):
    if x == '0-17':
        return 15
    elif x == '18-25':
        return 21
    elif x == '26-35':
        return 30
    elif x == '36-45':
        return 40
    elif x == '46-50':
        return 48
    elif x == '51-55':
        return 53
    else:
        return 60
df['Age'] = df['Age'].apply(age)
```

```python
df['Purchase'] = df['Purchase'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].replace('4+',4)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python
df['Product_ID'] = df['Product_ID'].astype(int)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df['City_Category'] = le.fit_transform(df['City_Category'])
```

```python
df.head()
```

```python
df1 = df[df['Purchase'] != 0]
df2 = df[df['Purchase'] == 0]
```

```python
df1.shape
```

```python
df2 = df2.drop('Purchase', axis = 1)
df2.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
a = sc.fit_transform(df1.drop('Purchase', axis = 1))
scaled_train = pd.DataFrame(a, columns= df1.columns[df1.columns != 'Purchase'])
scaled_test = pd.DataFrame(sc.transform(df2), columns= df2.columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, BaggingRegressor, AdaBoostRegressor
lr = LinearRegression()
bagg = BaggingRegressor()
rf = RandomForestRegressor()
gbm = GradientBoostingRegressor()
ada = AdaBoostRegressor()
```

```python
lr_pred = lr.fit(scaled_train, df1.Purchase).predict(scaled_test)
pred_bagg = bagg.fit(scaled_train, df1.Purchase).predict(scaled_test)
pred_rf = rf.fit(scaled_train, df1.Purchase).predict(scaled_test)
pred_gbm = gbm.fit(scaled_train, df1.Purchase).predict(scaled_test)
pred_ada = ada.fit(scaled_train, df1.Purchase).predict(scaled_test)
```

```python
solution = pd.DataFrame({'Purchase': lr_pred, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('LrPred.csv', index = False)
# 4694
```

```python
solution = pd.DataFrame({'Purchase': pred_bagg, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('Bagg.csv', index = False)
2861
```

```python
solution = pd.DataFrame({'Purchase': pred_rf, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('RF.csv', index = False)
## 2743
```

```python
solution = pd.DataFrame({'Purchase': pred_gbm, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('GBM.csv', index = False)
## 2932
```

```python
solution = pd.DataFrame({'Purchase': pred_ada, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('Ada.csv', index = False)
## 3769
```

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start = 1, stop = 12, num = 1)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
min_samples_split = [3,4,5,6,7]
min_samples_leaf = [1, 2, 5, 10]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2,
                               random_state=42, n_jobs = 1)
rf_random.fit(scaled_train, df1.Purchase)
```

```python
rf_random.best_estimator_
```

```python
rf1 = RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=15, max_features='auto', max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=10,
                      min_samples_split=3, min_weight_fraction_leaf=0.0,
                      n_estimators=1, n_jobs=None, oob_score=False,
                      random_state=None, verbose=0, warm_start=False)
pred_rf1 = rf1.fit(scaled_train, df1.Purchase).predict(scaled_test)
```

```python
solution = pd.DataFrame({'Purchase': pred_rf1, 'User_ID':test.User_ID, 'Product_ID':test.Product_ID})
solution.to_csv('RF1.csv', index = False)
## 2833
```

```python
from sklearn.ensemble import VotingRegressor

vc = VotingRegressor(estimators=[('gbm', gbm), ('rf', rf), ('bagg', bagg), ('ada', ada)])#voting='soft'

pred_vc = vc.fit(scaled_train, df1.Purchase).predict(scaled_test)
```
