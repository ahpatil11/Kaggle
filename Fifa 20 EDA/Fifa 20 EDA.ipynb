```python
df1 = df.drop(['sofifa_id', 'player_url', 'long_name', 'real_face', 'player_tags', 'team_jersey_number', 'loaned_from',
               'joined', 'ls','st','rs','lw','lf','cf','rf','rw','lam','cam','ram','lm','lcm','cm','rcm','rm','lwb','ldm',
               'cdm','rdm','rwb','lb','lcb','cb','rcb','rb', 'dob'],
        axis = 1)
```

# Top 10 Valuable Clubs

```python
a = df1.groupby('club').sum()['value_eur']
young = pd.DataFrame({'Value_of_the_club': a})
young = young.reset_index()
```

```python
c = young.sort_values('Value_of_the_club', ascending= False).head(10)
```

```python
plt.figure(figsize= (25, 5))
sns.barplot(c['club'], c['Value_of_the_club'])
```

# Top 10 Wages Club

```python
a = df1.groupby('club').sum()['wage_eur']
young = pd.DataFrame({'Wage_of_the_club': a})
young = young.reset_index()

b = young.sort_values('Wage_of_the_club', ascending= False).head(10)
```

```python
plt.figure(figsize= (25, 5))
sns.barplot(b['club'], b['Wage_of_the_club'])
```

# Top 10 Youngest Age Squad Club

```python
a = df1.groupby('club').mean()['age']
young = pd.DataFrame({'age': a})
young = young.reset_index()
```

```python
c = young.sort_values('age', ascending= True).head(10)
plt.figure(figsize= (25, 5))
sns.barplot(c['club'], c['age'])
```

# Top 10 Oldest Age Squad Club

```python
c = young.sort_values('age', ascending = False).head(10)
plt.figure(figsize= (25, 5))
sns.barplot(c['club'], c['age'])
```
